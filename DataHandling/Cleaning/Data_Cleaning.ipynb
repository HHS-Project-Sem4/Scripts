### Summer products

```python
summer_p_original = pd.read_csv("C:/Users/Robin/Documents/School/Project/DATA/EXTERNAL_DATASETS/SUMMER_PRODUCTS_summer-products-with-rating-and-performance_2020-08.csv")
# summer_p_profile = ProfileReport(summer_p_original, title="Profiling Report")
# summer_p_profile
```

#### duplicate rows

```python
summer_p = summer_p_original.copy()
summer_p['has_urgency_banner'] = summer_p['has_urgency_banner'].fillna(0)
summer_p['urgency_text'] = summer_p['urgency_text'].fillna('n.a.')
summer_p['merchant_profile_picture'] = summer_p['merchant_profile_picture'].fillna('n.a.')
summer_p[['rating_five_count' , 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']] = summer_p[['rating_five_count' , 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']].fillna(0)
summer_p[['product_color', 'product_variation_size_id', 'origin_country', 'merchant_name', 'merchant_info_subtitle']] = summer_p[['product_color', 'product_variation_size_id', 'origin_country', 'merchant_name', 'merchant_info_subtitle']].fillna('unknown')
summer_p = summer_p.drop_duplicates()
```
